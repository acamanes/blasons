# Exemple complet
On présente ci-dessous un exemple plus détaillé avec quelques fonctions cachées que nous n'avons pas vues précédemment.

Les phrases après des `#` sont des commentaires.

```python
from ecu import *

# Définition du blason
monblase = Ecu(forme = "moderne", email = ["or"])

# Partition du blason
b1, b2, b3, b4 = monblase.partition(
    nom = "écartelé",
    emaux = ["or","gueules",
           "gueules","or"]
)

# Description de la première partition
b1.piece(nom="chef", email=["azur"])
b1.meuble_piece(piece="chef",
                nom="fleur de lys",
                nombre = 5,
                emaux=["or"])
b1.meuble(nom="aigle au vol abaissé",
          emaux=["sable","gueules",
                 "or","gueules","argent"])

# La deuxième partition est elle-même partitionnée
b21, b22, b23, b24 = b2.partition(
    nom = "écartelé",
    emaux=["or", "gueules","gueules","or"])
## Chacune des partitions est elle-même un blason...
b21.partition(nom = "fascé",emaux=["azur","sinople"])
b22.piece(nom = "chef",email=["argent"])
b22.meuble_piece(piece="chef",nom="ancre",
                 nombre=4,
                 emaux=["gueules"])
b22.meuble(nom="rat",
           translation=(0,-b22.height/6),
        emaux=["argent","sable"])

b23.meuble(nom="cerbère",
           emaux=["argent","or","or","sable"])

b24.partition(nom = "écartelé",
              emaux = ["azur","sinople","sinople","azur"])

# La troisième partition : des meubles un peu améliorés
b3.meuble(nom="lion rampant",
          contourne=False,
          translation=(b3.width/5,0),
          echelle=1,
          emaux=["or", "gueules", "azur", "argent"])
b3.meuble(nom="dauphin",
          contourne=True,
          translation=(-b3.width/5,0),
          echelle=1,
          emaux=["argent", "or", "argent"])

# La quatrième partition est partitionnée
b41, b42, b43, b44 = b4.partition(
    nom = "écartelé",
    emaux=["gueules","or","or","gueules"])
b41.piece(nom = "4 burelles",
          email=["argent"])
b42.partition(nom = "taillé",
              emaux = ["sable","argent"])
b43.piece(nom = "2 barres",
          email = ["gueules"])
b44.partition(
    nom = "écartelé",
    emaux = ["or","argent","argent","or"])
b44.meuble(nom="tortue",
           emaux=["sinople","sable",
                  "sinople","argent"])

# Enregistrement du fichier avec possibilite de choisir le nom
monblase.dessine(fichier="alain_blase")
```
