# Création d'un blason simple
Chaque case grisée peut être modifiée. Chacune de ces cases est précédée d'une courte explication. Une fois toutes les cases grisées modifées pour décrire le blason que vous souhaitez, il suffit de cliquer sur le bouton Run **à modifier pour avoir restart** en haut de la page pour lancer la création du blason qui s'affiche en fin de page et s'enregistre aux formats pdf et png.

On commence par importer les fonctions qui permettent de réaliser le blason.

```python
from ecu import *
```

## Création du blason
La fonction suivante permet de créer un blason.

Vous pouvez modifier la forme du blason en choisissant entre 
- "ancien" 
- "moderne"

**Attention** Le nom est entre guillemets et son orthographe doit être respectée

Vous pouvez modifier l'émail du blason en choisissant parmi 
- les métaux : "or", "argent"
- les émaux : "gueules" (rouge), "azur" (bleu), "sable" (noir), "sinople" (vert), "pourpre" (violet)

Vous pouvez changer le nom dans lequel est stocké le blason (ici `monblase`), à condition d'utiliser le même nom par la suite.

```python
monblase = Ecu(
    forme = "moderne",
    email = ["azur"]
)
```

## Ajout d'un meuble
Sur ce blason, vous pouvez ajouter un meuble. La liste des meubles est disponible [ici](presentation_meubles.pdf).

Pour chacun des meubles, il faut préciser
- son `nom` : disponible au-dessus de l'image. Respectez les accents et les espaces
- ses `emaux` : chacun des meubles est divisé en plusieurs caractéristiques (griffes, bec, langue,...). La liste des caractéristiques est donnée avec la liste des meubles. Les émaux doivent être donnés sous forme de liste. Le premier émail correspond à la première caractéristique, le deuxième à la deuxième,...

**Attention** si vous avez modifié le nom de votre blason, il faut également le modifier ci-dessous.

```python
monblase.meuble(
    nom = "aigle au vol abaissé",
    emaux = ["or", "argent", "argent", "gueules", "azur"])
```

## Création du blason
La commande suivante permet de créer le blason. Il s'affiche en-dessous et s'enregister sous format pdf et png dans votre dossier de travail. Vous devez choisir le nom du `fichier`.

**Attention** le nom du fichier **ne** doit **pas** contenir d'accent et d'espaces.

```python
monblase.dessine(fichier = "moi_blason_simple")
```
