# Blasons et Partitions
Le besoin de créer de plus en plus de blasons différents, l'envie de représenter dessus des alliances conduit à créer des blasons avec plusieurs parties.

Commençons par définir un écu, comme nous en avons l'habitude.

```python
from ecu import *
monblase = Ecu(forme = "moderne", email = ["argent"])
```

## Partitionnement
Le partitionnement consiste à découper un blason en plusieurs blasons plus petits.

La liste des partitions est disponible [ici](presentation_partitions.pdf)

La **partition** crée des petits blasons. Comme on le voit ci-desous, chacun des petits blasons est stocké dans une nouvelle variable. À gauche du signe `=` il doit donc y avoir autant de noms (séparés de virgules) qu'il y a de petits blasons créés. Ici, on construit un blason **écartelé** donc on crée **4** petits blasons.

Les **émaux** doivent être écrits dans l'ordre de la partition. Cet ordre est précisé dans le fichier pdf précédent, pour les partitions les plus réduites.

```python
b1, b2, b3, b4 = monblase.partition(
    nom = "écartelé",
    emaux = ["gueules", "or", "or", "gueules"])
```

# Description des partitions
Les blasons `b1`, `b2`, `b3`, `b4` définis précédemments sont des blasons... on peut donc leur appliquer ce que nous avons vu dans les tutos précédents.

```python
b1.meuble(nom = "aigle au vol abaissé", emaux = ["or", "argent", "argent", "argent", "azur"])
b2.piece(nom = "chef", email = ["azur"])
b2.meuble_piece(nom = "fleur de lys", emaux = ["or"], nombre = 5)
b2.meuble(nom = "rat", emaux = ["sable", "argent"])
b3.piece(nom = "bande", email = ["sinople"])
b4.meuble(nom = "marmitte", emaux = ["or", "argent"])
```

# Création du blason
Comme précédemment on enregistre le *grand* blason. C'est pourquoi, avant `.dessine` on utilise la variable `monblase` définie au tout début.

```python
monblase.dessine(fichier="tuto_partition")
```
